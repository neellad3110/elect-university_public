```python
name_list=list()
rank_list=list()
logo_list=list()
link_list=list()
country_list=list()
region_list=list()
flag=list()
foundation_year_list=list()
address_list=list()
website_list=list()
intro_list=list()
total_enrollment_list=list()
international_total_enrollment_list=list()
ug_enrollment_list=list()
ug_international_enrollment_list=list()
pg_enrollment_list=list()
pg_international_enrollment_list=list()
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import cssutils
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time


driver = webdriver.Chrome()
inner_driver = webdriver.Chrome()

url="https://www.shanghairanking.com"


driver.get('https://www.shanghairanking.com/rankings/arwu/2023')

# # Wait for the dynamic content to load (you may need to adjust the waiting time)
# driver.implicitly_wait(10)

# # Get the page source after content has loaded
# page_details=driver.page_source

page_counter=1
while True:
    # Wait for the dynamic content to load (you may need to adjust the waiting time)
    driver.implicitly_wait(10)

    # Get the page source after content has loaded
    page_details=driver.page_source
    # Parsing the HTML
    soup = BeautifulSoup(page_details, 'html.parser')

    ranktable = soup.find('table', class_='rk-table')
    tablebody = ranktable.find('tbody')
    tabledetail=tablebody.find_all('tr')


    for i in tabledetail:
        
        info_list=list()
        keystats_list=list()

        if page_counter>=13:
            rank= i.find('div', class_='ranking')    
            if(rank):
                print("rank :",rank.text.strip())
                rank_list.append(rank.text.strip())

                headingdiv=i.find('div',class_="global-univ")
                logo=url+headingdiv.find('img',class_="univ-logo").get('src').strip()
                print("logo :",logo)
                logo_list.append(logo)

                link=headingdiv.find('a').get('href')
                if(link):
                    print("link :",url+link)
                    link_list.append(url+link)
                
                name=headingdiv.find('span',class_="univ-name").text.strip()
                print("name :",name)
                name_list.append(name)


                flagdiv=i.find('div',class_="region-img")['style']
                flagstyle = cssutils.parseStyle(flagdiv)
                
                flagurl = flagstyle['background-image']
                flagurl=flagurl.replace('url(','').replace(')','')
                print("flagurl :",url+flagurl.strip())
                flag.append(url+flagurl)
                
                

                inner_driver.get(url+link);
                inner_driver.implicitly_wait(10)
                inner_page_source=inner_driver.page_source
                innersoup = BeautifulSoup(inner_page_source, 'html.parser')
                
                contactdiv=innersoup.find('div',class_="contact-msg")
                info=contactdiv.find_all('div',class_="contact-item")
                
                for j in info:
                    info_list.append(j.text.strip().split(':',1)[1])

                if(innersoup.find('div',id="univ_intro")):
                    introtext=innersoup.find('div',id="univ_intro").text
                    info_list.append(introtext.strip().split(' ',1)[1])
                else:
                    info_list.append('NA')
                    
                if info_list[0] : 
                    print("region :",info_list[0])
                    region_list.append(info_list[0])
                else : 
                    print("region :",info_list[0])
                    region_list.append('NA')   

                print("country :",info_list[1])
                country_list.append(info_list[1]) 

                if(info_list[2]):
                    print("year :",info_list[2])
                    foundation_year_list.append(info_list[2]) 
                else :
                    print("year :",info_list[2])
                    foundation_year_list.append('NA') 

                if(info_list[3]):
                    print("address :",info_list[3])   
                    address_list.append(info_list[3])
                else :
                    print("address :",info_list[3])   
                    address_list.append('NA')    

                if(info_list[4]):
                    print("website :",info_list[4])
                    website_list.append(info_list[4])
                else:
                    print("website :",info_list[4])
                    website_list.append('NA')

                if(info_list[5]):    
                    print("Intro :",info_list[5])
                    intro_list.append(info_list[5])
                else :
                    print("Intro :",info_list[5])
                    intro_list.append("NA")    

                

                statsdiv=innersoup.find_all('div',class_="box-container")
                
                for k in statsdiv: 

                    statsrow=k.find_all('div',class_="item")
                    for l in statsrow:
                        statsvalue=l.find('div',class_="num-box").text
                        keystats_list.append(statsvalue.strip())

                
                print("Total Enrollment :",keystats_list[0])
                total_enrollment_list.append(keystats_list[0])

                print("Total International Enrollment :",keystats_list[1])
                international_total_enrollment_list.append(keystats_list[1])

                print("UG Enrollment :",keystats_list[2])
                ug_enrollment_list.append(keystats_list[2])

                print("International UG Enrollment :",keystats_list[3])
                ug_international_enrollment_list.append(keystats_list[3])

                print("PG Enrollment :",keystats_list[4])
                pg_enrollment_list.append(keystats_list[4])

                print("International PG Enrollment :",keystats_list[5])
                pg_international_enrollment_list.append(keystats_list[5])

                print("-------------------------------------------------");
    
  
    page_counter=page_counter+1;

    if(page_counter>5 and page_counter<=31):
        next_button = driver.find_element(By.XPATH,"(//a[@class='ant-pagination-item-link'])[4]")
    else:
        next_button = driver.find_element(By.XPATH,"(//a[@class='ant-pagination-item-link'])[3]")

    if next_button and page_counter<=34:
        next_button.click()
        time.sleep(2)  # Add a delay to allow the page to load
    else:
        break
    
    
        
# Close the Selenium WebDriver
driver.quit()
inner_driver.quit()


```

```python

import pandas as pd

df = pd.DataFrame(list(zip(rank_list,logo_list,name_list,link_list,website_list,flag,country_list,region_list,foundation_year_list,address_list,intro_list,total_enrollment_list,international_total_enrollment_list,ug_enrollment_list,ug_international_enrollment_list,pg_enrollment_list,pg_international_enrollment_list)),
               columns =['Rank', 'Logo','Name','Link','Website','Flag','Country','Region','Foundation_year','Address','Intro','Total_Enrollment','Total_International_Enrollment','UG_Enrollment','International_UG_Enrollment','PG_Enrollment','International_PG_Enrollment'])
df
```
